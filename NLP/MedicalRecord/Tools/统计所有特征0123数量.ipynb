```python
indata_path = '测试_全特征.txt'
```

```python
cols, data = [], []
with open(indata_path) as f:
    for idx, line in enumerate(f.readlines()):
        arr = line.strip().split('	')[3:]
        if idx == 0:
            cols = arr
            data = [{'0': 0, '1': 0, '2': 0, '3':0} for c in cols]
        else:
            for e_idx, e in enumerate(arr):
                if e_idx == 30:
                    continue
                data[e_idx][e] = data[e_idx][e] + 1

for col, col_dict in zip(cols, data):
    print('%s	%s	%s	%s	%s' % (col, col_dict['0'], col_dict['1'], col_dict['2'], col_dict['3']))
            
        
```
