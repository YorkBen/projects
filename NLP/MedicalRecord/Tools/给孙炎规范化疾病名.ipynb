```python
#### 暂时用不上
def restate_dict(data, num):
    t_dict, map_dict = {}, {}
    for k, v in data:
        if v >= num:
            t_dict[k] = v
        else:
            break
            
    for k, v in data:
        if v < num:
            for k2 in t_dict.keys():
                if k2 in k:
                    map_dict[k] = k2
                    
    return t_dict, map_dict
```

```python
import re

def replace_line(line):
    line = line.strip()
    s1 = line.replace('至今', '').replace('已治愈', '').replace('摔伤致', '').replace('病史', '').replace('已切除', '')
    s1 = s1.replace('诊断得', '').replace('UK.UK-', 'UK').replace('行肠镜检查', '').replace('确诊', '').replace('已行', '')
    s1 = s1.replace('目前持续情况不明', '').replace('目前病情稳定', '').replace('现无用药', '').replace('无用药', '')
    s1 = s1.replace('自觉', '').replace('偶有', '').replace('出现', '').replace('现已恢复', '').replace('其他的', '')
    s1 = s1.replace('不伴有', '，').replace('不伴', '，').replace('伴有', '，').replace('伴', '，')
    return s1


exclude_signs = ['（C2）', 'CA', '十二指', 'TA724', 'CT724', '0-IIb', '（3级）', '半月板']
def has_exclude_signs(s):
    for e in exclude_signs:
        if e in s:
            return True
    return False


# 读数据
data3, line_data = [], []
with open('t3.txt') as f:
    for line in f.readlines():
        row_data = []
        arr = re.split('[，,、；。]', replace_line(line))
        for s in arr:
            if len(s) < 2:
                continue
            else:
                if not has_exclude_signs(s):
                    s1 = re.sub('[0-9几两个年月日天周有余十九八七六五四三二一半.患者于初步诊断\-（）提示：]{2,}', ' ', s)
                    s1 = re.sub('.[年周月天日]左右', '', s1)
                    s1 = s1.replace('发现', '').replace('胃镜', '').replace('超声', '').replace('CT', '')
                    s1 = s1.replace('外院', '').replace('检查', '').replace('液深约cm', '').replace('诊断', '')
                    if s != s1:
                        print(s, '-----》',s1)
                        arr2 = s1.split(' ')
                        for e2 in arr2:
                            if len(e2) >= 2:
                                data3.append(e2.strip())
                                row_data.append(e2.strip())
                    else:
                        data3.append(s.strip())
                        row_data.append(s.strip())
                else:
                    print('！！！特殊符号', s)
                    data3.append(s.strip())
                    row_data.append(s.strip())
        line_data.append(row_data)
        
# 统计
diease_dict = {}
for e in data3:
    if e not in diease_dict:
        diease_dict[e] = 0
    diease_dict[e] = diease_dict[e] + 1

# 排序
data33 = [(k, v) for k,v in diease_dict.items()]
data33 = sorted(data33, key=lambda x: x[1], reverse=True)
```
