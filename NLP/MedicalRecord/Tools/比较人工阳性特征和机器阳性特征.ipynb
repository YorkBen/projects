```python
data1 = []
with open('t1.txt') as f:
    for line in f.readlines():
        s1 = []
        arr = line.strip().split('	')
        for e in arr[2:]:
            if e != '':
                s1.append(e)
        data1.append(set(s1))
        
data2 = []
with open('t2.txt') as f:
    for line in f.readlines():
        s2 = []
        arr = line.strip().split('	')
        for e in arr:
            if e != '':
                s2.append(e)
        data2.append(set(s2))
        
```

```python
for s1, s2 in zip(data1, data2):
    print(s1 - s2, '	',s2 - s1)
```
