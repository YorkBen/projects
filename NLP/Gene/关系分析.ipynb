```python
file_path = r'Data/positive.txt'
def get_relation_text(file_path):
    data = []
    with open(file_path) as f:
        for line in f.readlines():
            arr = line.strip().split('	')
            label = arr[-1]
            for text in arr[0].split('. '):
                if '#' in text and '$' in text:
#                     data.append((text.replace('#', '').replace('$', ''), label))
                    data.append((text, label))
            
    return data
```

```python
text1 = get_relation_text(r'Data/positive.txt')
# text2 = get_relation_text(r'Data/negative.txt')
# texts = list(set(text1 + text2))
texts = list(set(text1))
```

```python
for txt, label in texts:
    print(txt, label)
    print('')
```

```python
import spacy
from spacy import displacy
from spacy.symbols import ORTH
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
special_case = [{ORTH: "β-catenin"}]
nlp.tokenizer.add_special_case("β-catenin", special_case)
```

```python
doc = nlp("Wnt4/β-catenin signal pathway was inactivated by inhibiting the expression of circ_PVT1.")
```

```python
displacy.serve(doc, style="dep")
```
