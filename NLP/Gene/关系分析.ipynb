```python
file_path = r'Data/positive.txt'
def get_relation_text(file_path):
    data = []
    with open(file_path) as f:
        for line in f.readlines():
            arr = line.strip().split('	')
            label = arr[-1]
            for text in arr[0].split('. '):
                if '#' in text and '$' in text:
#                     data.append((text.replace('#', '').replace('$', ''), label))
                    data.append((text, label))
            
    return data
```
