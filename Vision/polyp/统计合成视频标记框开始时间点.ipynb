```python
import os
```

```python
delta = 4
fps = 20
base_dir=r'D:\projects\Vision\polyp\data'
folders = ['2008_0.03', '2008_0.1', '2008_0.35', '2010_0.05', '2010_0.2', '2010_0.35', \
           '2016_0.05', '2016_0.15', '2016_0.35', '2019_0.02', '2019_0.03', '2019_0.1', \
           '3006_0.05', '3006_0.1', '3006_0.35']
```

```python
def get_dir_sorted_imgs(dirs):
    names = []
    if isinstance(dirs, str):
        if os.path.exists(dirs):
            names = [int(filename[:-4]) for filename in os.listdir(dirs) if filename.endswith('.jpg')]
    elif isinstance(dirs, list):
        for dir in dirs:
            if os.path.exists(dir):
                names.extend([int(filename[:-4]) for filename in os.listdir(dir) if filename.endswith('.jpg')])
    names = sorted(names)
    img_names = ['%s.jpg' % name for name in names]
    return img_names
```

```python
def stat_ct_by_imgNames(folder):
    arr = folder.split('_')
    video_name, score = arr[0], arr[1]
    input_img_path1 = os.path.join(base_dir, r'input\imgs\%s_crop_select' % video_name)
    input_img_path2 = os.path.join(base_dir, r'input\imgs\%s_crop_select2' % video_name)
    input_img_path = os.path.join(base_dir, r'input\imgs\%s_crop' % video_name)
    process_img_path = os.path.join(base_dir, r'output\imgs\%s_%s' % (video_name, score))
    
#     img_names = get_dir_sorted_imgs([input_img_path1, input_img_path2])
    img_names = get_dir_sorted_imgs(input_img_path)
    img_names_processed = get_dir_sorted_imgs(process_img_path)
    
    print(len(img_names))
    print(len(img_names_processed))
    
    results,start,end = [], -1, -1
    for name in img_names_processed:
        pos = img_names.index(name)
#         print(name, pos, start, end, delta)
        if start == -1:
            start = pos
            end = pos
        else:
            if pos < end + delta:
                end = pos
#                 print('update end')
            else:
                if end - start >= delta:
                    results.append([start, end])
#                     print('append start end')
                start = pos
                end = pos

    return results
```

```python
stat_ct_by_imgNames('2008_0.03')
```

```python
for folder in folders:
    results = stat_ct_by_imgNames(folder)
    print(folder)
    for r in results:
        s = r[0] / fps
        t = '%d%.3f' % (s // 60, s % 60)
        print(t)
```
